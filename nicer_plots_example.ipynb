```python
amp_par0 = c0.get_param('B1937+21_dm_gp_gamma', to_burn=True)
gam_par0 = c0.get_param('B1937+21_dm_gp_log10_A', to_burn=True)
amp_par1 = c1.get_param('B1937+21_dm_gp_gamma', to_burn=True)
gam_par1 = c1.get_param('B1937+21_dm_gp_log10_A', to_burn=True)

#Here is the script to get the peak of the 2d histogram
log10_A0, gamma0 = utils.get_params_2d_mlv(c0,
                                           'B1937+21_dm_gp_gamma',
                                           'B1937+21_dm_gp_log10_A')
log10_A1, gamma1 = utils.get_params_2d_mlv(c1,
                                           'B1937+21_dm_gp_gamma',
                                           'B1937+21_dm_gp_log10_A')
```

```python
fig = plt.figure(figsize=[6,5])
axis = fig.add_subplot(111)
c.hist2d(amp_par0,gam_par0,
         bins=40, ax=axis, 
         plot_datapoints=False,
         plot_density=False,
         plot_contours=True,
         no_fill_contours=True, color='C0',
         levels=(0.68,0.95))
c.hist2d(amp_par1,gam_par1,
         bins=40, ax=axis, 
         plot_datapoints=False,
         plot_density=False,
         plot_contours=True,
         no_fill_contours=True, color='C1',
         levels=(0.68,0.95))

#This plots the "center" of the 2d histogram
plt.plot(log10_A0, gamma0,'x',markersize=15,color='k')
plt.plot(log10_A1, gamma1,'+',markersize=15,color='k')

plt.title('Comparing Chromatic Red Noise')
# plt.legend()
plt.xlim(0,5)
plt.ylim(-14,-11)
lines=[plt.Line2D([0], [0],color='C0',linewidth=2),
       plt.Line2D([0], [0],color='C1',linewidth=2)]
labels = ['NG Radio Only', 'NG + NICER']
plt.legend(lines,labels,loc='lower left',fontsize=12)
plt.show()
```
